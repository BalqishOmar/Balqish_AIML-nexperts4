```python
# Read the file 

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

path = r"C:\Users\Aqilah Othman\OneDrive\Desktop\Aqilah\Project Dashboard\ramen-ratings.csv"

df_ramen = pd.read_csv(path)

print(df_ramen)
df_ramen.head()
```

```python
# Data Cleaning

# Check data type
print(df_ramen.dtypes)
print("\nUnique values in Stars:", df_ramen['Stars'].unique())
```

```python
# Remove 'Unrated' rows and convert Stars to numbers
df_ramen = df_ramen[df_ramen['Stars'] != 'Unrated']
df_ramen['Stars'] = pd.to_numeric(df_ramen['Stars'])

# Confirm it worked
print(df_ramen.dtypes)
print(f"Remaining rows: {len(df_ramen)}")
```

```python
# Check for missing values
print(df_ramen.isnull().sum())
```

```python
# Check for missing values
print(df_ramen.isnull().sum())
```

```python
# Chart 1 : Top 10 Countries by Number of Ramen Reviews

plt.figure(figsize=(10, 6))

country_counts = df_ramen['Country'].value_counts().head(10)
colors = ['#4361ee', '#4895ef','#4cc9f0', '#3f37c9', '#560bad',
          '#7209b7', '#b5179e', '#f72585', '#3a0ca3', '#480ca8']

sns.barplot(x=country_counts.values, y=country_counts.index, palette = colors)
plt.title('Top 10 Countries by Reviews', fontsize = 13, fontweight='bold')
plt.xlabel('Number of Reviews')
plt.ylabel('Country')
plt.tight_layout()
plt.savefig('Chart1.png', dpi=150, bbox_inches='tight')
plt.show()
```

```python
# Chart 2 : Average Star Rating by Country (Top10)

plt.figure(figsize=(10, 6))
top_countries = df_ramen['Country'].value_counts().head(10).index
country_avg = df_ramen[df_ramen['Country'].isin(top_countries)].groupby('Country')['Stars'].mean().sort_values()

plt.plot(country_avg.index, country_avg.values, color='#4361ee', linewidth=2.5, marker='o', markersize=8)
plt.fill_between(country_avg.index, country_avg.values, alpha=0.1, color='#4361ee')
plt.title('Average Rating by Top 10 Countries', fontsize=13, fontweight='bold')
plt.xlabel('Country')
plt.ylabel('Average Stars')
plt.ylim(3.0, 4.5)
plt.xticks(rotation=30)
plt.grid(axis='y', linestyle='--', alpha=0.5)
plt.tight_layout()
plt.savefig('Chart2.png', dpi=150, bbox_inches='tight')
plt.show()
```

```python
# Chart 3 : Ramen Style Distribution

plt.figure(figsize=(10, 8))
style_counts = df_ramen['Style'].value_counts()
custom_order = ['Pack', 'Tray', 'Box', 'Bowl', 'Can', 'Cup', 'Bar']
style_reordered = style_counts.reindex(custom_order).dropna()

colors = ['#4361ee', '#4895ef', '#4cc9f0', '#3f37c9', '#560bad', '#7209b7', '#b5179e']

plt.pie(
    style_reordered.values,
    labels=style_reordered.index,
    autopct='%1.1f%%',
    colors=colors,
    startangle=140,
    pctdistance=0.75,
    labeldistance=1.15
)
plt.title('Ramen Style Distribution', fontsize=13, fontweight='bold')
plt.tight_layout()
plt.savefig('Chart3.png', dpi=150, bbox_inches='tight')
plt.show()
```

```python
import matplotlib.patches as mpatches

plt.figure(figsize=(10, 6))
brand_stats = df_ramen.groupby(['Brand', 'Country']).agg(
    total_reviews=('Stars', 'count'),
    avg_rating=('Stars', 'mean')
).reset_index()
brand_stats = brand_stats[brand_stats['total_reviews'] > 5]

# Use only top 10 countries for cleaner legend
top10_countries = df_ramen['Country'].value_counts().head(10).index
brand_stats = brand_stats[brand_stats['Country'].isin(top10_countries)]

colors_list = ['#4361ee','#4895ef','#4cc9f0','#3f37c9','#560bad',
               '#7209b7','#b5179e','#f72585','#3a0ca3','#480ca8']
countries = top10_countries
color_map = dict(zip(countries, colors_list))
colors = brand_stats['Country'].map(color_map)

plt.scatter(brand_stats['total_reviews'], brand_stats['avg_rating'],
            c=colors, s=80, edgecolors='white', linewidth=0.5, alpha=0.85)

# Label top 5
top5 = brand_stats.nlargest(5, 'total_reviews')
for _, row in top5.iterrows():
    plt.annotate(row['Brand'], (row['total_reviews'], row['avg_rating']),
                textcoords="offset points", xytext=(-40, 8), fontsize=8, color='#333')

legend_handles = [mpatches.Patch(color=color_map[c], label=c) for c in countries]
plt.legend(handles=legend_handles, bbox_to_anchor=(1.01, 1), loc='upper left',
           fontsize=7, title='Country', title_fontsize=8)

plt.title('Brand Popularity vs Average Rating', fontsize=13, fontweight='bold')
plt.xlabel('Number of Reviews')
plt.ylabel('Average Stars')
plt.grid(linestyle='--', alpha=0.4)
plt.tight_layout()
plt.savefig('Chart4.png', dpi=150, bbox_inches='tight')
plt.show()
```
