# $$Visualizations$$

- What we are doing is called Exploratory Data Analysis (EDA)
- Consists of 2 main steps:
    1. Data Prepration/Preprocessing - 
    2. Data Visualization - 

- We can do the following things:
    1. identify relationships between variables
    2. uncover hidden patterns
    3. spot any anomalies and outliers
    4. prepare data for the further analysis or modeling

TYPES:
1. Univariate Analysis:
    - involves a single variable

2. Bivariate Analysis:
    - involves two variables

3. Multivariate Analysis:
    - involves a multiple variables at the same time

##### Bar Graph
- Two Variables - Bi variate analysis
- A bar graph has bars, based on the count/magnitude of each category in the graph
    - need to have to catagorical data
    - 
-  you have categories on x-axis
- magnitude on y-axis

```python
cities = ['Kuala Lumpur','Honk Kong','New York','Singapore','California']
temperature = [28,32,11,42,23]

```

```python
import matplotlib.pyplot as plt

plt.bar(cities, temperature, color = 'skyblue')
```

```python
plt.figure(figsize= (8,3))
plt.bar(cities, temperature, 
        color = 'skyblue',
        width=0.7, # thickness of the bars
        
        edgecolor = "black", # is the boundary of bars
        linewidth = 1, # thickness of the edge

        alpha = 0.8, # transparency
        hatch = ".xOo-|" # / \ x X O o | + . * - # adds patterns to a graph
        )
```

```python
colors = ['skyblue', 'lightgreen', 'lightyellow', 'lightpink', 'orange']
hatches = ['X', '//', '|\\', 'Oo', '+o']
edgecolors = ['blue', 'green', 'yellow', 'red', 'black']

plt.bar(cities, temperature, color = colors, hatch = hatches, edgecolor = edgecolors)

plt.title("Temperature of 5 Cities")
plt.xlabel("-----------Cities------------")
plt.ylabel("-------------Temperatures------------")

plt.legend()
plt.grid()

plt.show()
```

```python
import numpy as np

data = np.random.randint(1,10,1000) # random list of numbers # 1 thousand elements

data = list(range(10)) + list(range(20)) + list(range(5))

plt.hist(data,
        bins= 3,
        color='skyblue',
        edgecolor = 'black',
        alpha = 0.5
        )

plt.title("HISTOGRAM --- Frequency counter")
plt.xlabel("-----------Variable------------")
plt.ylabel("-------------Frequency------------")
plt.show()
```

##### Scatter Plot
- bivariate

```python
plt.scatter(cities, temperature,
            s = 500, # size of the points / bubbles
            # c = colors, # control of size of the points/bubbles
            marker= '*', # ^-triangle s-square x-cross +_plus  *-star D- rhombus
            cmap= "viridis", vmin = 0, vmax = 40, c = temperature
            )


plt.title("Temperature of 5 Cities")
plt.xlabel("-----------Cities------------")
plt.ylabel("-------------Temperatures------------")
plt.show()
```

##### Pie Chart

```python
plt.pie(temperature,
        labels = cities, 
        autopct= '%.1f%%', # list the percentage --- .2f for precision upto two places
        startangle= 90,
        wedgeprops= {"width": 0.6}, # use this to get a donut chart
        colors= colors

        )

plt.title("Temperature of 5 Cities")
plt.xlabel("-----------just for some additional info------------")
# plt.ylabel("-------------Temperatures------------")
plt.show()
```
