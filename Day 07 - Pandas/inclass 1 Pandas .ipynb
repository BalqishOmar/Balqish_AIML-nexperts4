# $$Pandas$$

- Pandas is a python library --- `Panel Data`

Dimensions | Name | Description
----|---|---
1 | Series  | 1D   labeled homogeneously-typed array
2 | DataFrame | General 2D   labeled, size-mutable tabular structure with potentially heterogeneously-typed column

### Temporal Data - OPTIONAL

```python
from datetime import datetime

# -------------------------------
# 1. Creating a datetime column
# -------------------------------
# Example data with date strings
data = {
    'event': ['Start', 'Middle', 'End'],
    'date_str': ['2024-01-15', '2024-05-20', '2024-12-31']
}

df = pd.DataFrame(data)
```

```python
# Convert string to datetime
df['date'] = pd.to_datetime(df['date_str'], format='%Y-%m-%d', errors='coerce')

df.dtypes
```

```python
# -------------------------------
# 2. Extracting date components
# -------------------------------
df['year'] = df['date'].dt.year
df['month'] = df['date'].dt.month
df['day'] = df['date'].dt.day
df['weekday'] = df['date'].dt.day_name()
```

```python
# -------------------------------
# 3. Filtering by date range
# -------------------------------
start_date = '2024-05-01'
end_date = '2024-12-31'
mask = (df['date'] >= start_date) & (df['date'] <= end_date)
filtered_df = df.loc[mask]
```

```python
# -------------------------------
# 4. Current date & time
# -------------------------------
now = pd.Timestamp.now()
today = pd.Timestamp.today()
print("Current Timestamp:", now)
print("Today's Date:", today.date())
print(today)
print("Today's Date:", today.date())
```

```python
# -------------------------------
# 5. Date arithmetic
# -------------------------------
df['date_plus_7_days'] = df['date'] + pd.Timedelta(days=7)
df['days_since_event'] = (today - df['date']).dt.days
```
